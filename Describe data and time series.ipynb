# PM10

```python
import pandas as pd
data = pd.read_excel('PM10_process.xlsx')

data.describe()
```

```python
import matplotlib 
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
X = data['PM10']
times = pd.date_range(start='2017-03-21',end='2018-03-17',periods=362)
fig, ax = plt.subplots(figsize=(30, 12))
matplotlib.pyplot.plot_date(times,X,'r',linestyle ='solid',linewidth = 2)
plt.title('Time Series Of PM10 Concentration (Naderi Station)',fontsize=22)
plt.xlabel('Time (Day)',fontsize=22)
plt.ylabel('PM10 (Concentration)',fontsize=26)
plt.xticks(color='K', size=18)
plt.yticks(color='K', size=18)
x = mpatches.Patch(label='0-50 Good')
y = mpatches.Patch(label='51-100 Moderate ')
z = mpatches.Patch(label='101-200 Unhealthy ')
w = mpatches.Patch(label='201-300 very Unhealthy ')
U = mpatches.Patch(label='301-  Hazardous ')
plt.grid()
plt.legend(handles=[x, y,z,w,U],loc=2)
fig.savefig('PM10_finall.png', dpi=plt.gcf().dpi, bbox_inches = 'tight')
plt.show()
```
