```python
import pandas as pd
data = pd.read_excel('All_pollutant_standard.xlsx')
day = []
for i in range(1,337):
    day.append(i)
data['DAY'] = day
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
X1 = data[['PM10','PM2.5','NOx','Tempreture','Humid','Wind Speed','Wind Direction Sin','Wind Direction COS','Pressure','RainFall','DAY']]
X1 = X1.values
sc_X = StandardScaler()
X1_1 = sc_X.fit_transform(X1)
```

```python
import numpy as np
sin_file = np.array(X1_1)
df = pd.DataFrame(sin_file,columns=['PM10','PM2.5','NOx','Tempreture','Humid','Wind Speed','Wind Direction Sin','Wind Direction COS','Pressure','RainFall','DAY'])
```

```python
df.corr(method='pearson')
```
